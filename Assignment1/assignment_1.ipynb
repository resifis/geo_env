# Needed Packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pdb
import xarray as xr
```

# 1- Load data

```python
path_data = "N21E039.SRTMGL1_NC.nc" # path to the data
dset = xr.open_dataset(path_data) # read the data using xarray
#pdb.set_trace() # debugging step 
DEM = np.array(dset.variables["SRTMGL1_DEM"]) # extract the needed variable
dset.close()
#pdb.set_trace() # debugging step 
```

# 2- Visualize

```python
plt.figure(figsize = (8,5)) # define the size of the figure
plt.title("Visualization of Jeddah's Topography",fontsize = 20)
plt.imshow(DEM,cmap = "jet") # plot the extracted data
cbar = plt.colorbar(shrink = 0.7) #generate a colorbar
cbar.set_label("Elevation (m asl)") # colorbar text label
plt.savefig("assignment_1.png",dpi = 300) # save the generated figure
plt.show()
```

$\large \textbf{Description:}$

The figure above represents the elevation map of Jeddah's topography.
We can notice that the elevation is increasing from West to East indicating the presence of mountain in the Eastern part of the map. We can also see that the highest values of elevation are in the north-east part of the map.
